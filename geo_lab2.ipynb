```python
import pandas as pd
import geopandas as gpd

# Filepath
fp = "data/Helsinki/addresses.txt"

# Read the data
data = pd.read_csv(fp, sep=";")
data.head()
```

```python
from geopandas.tools import geocode

geo = geocode(
    data["addr"], provider="nominatim", user_agent="day88", timeout=10
)
geo.head()
```

```python
geo.shape
```

```python
join = geo.join(data)
join.head()
```

```python
join.shape
```

```python
join.explore(color="red", marker_kwds={"radius": 5})
```
